```
# Creating the authentication object
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
# Setting your access token and secret
auth.set_access_token(access_token, access_token_secret)
# Creating the API object while passing in auth information
api = tweepy.API(auth) 
```

```
api.rate_limit_status()
```

```
file_directory = "/content/gdrive/My Drive/Pandemic Spread - Social Media/Top Countries Data/dataset/"
```

```
def get_tweets(handle_name, country):
    print(handle_name, country)
    
    username = handle_name
    startDate = datetime.datetime(2019, 12, 1, 0, 0, 0)
    endDate =   datetime.datetime(2020, 4, 18, 0, 0, 0)

    tweets = []
    tmpTweets = api.user_timeline(username)
    for tweet in tmpTweets:
        if tweet.created_at < endDate and tweet.created_at > startDate:
            tweets.append(tweet)

    while (tmpTweets[-1].created_at > startDate):
        print("Last Tweet @", tmpTweets[-1].created_at, " - fetching some more")
        tmpTweets = api.user_timeline(username, max_id = tmpTweets[-1].id)
        for tweet in tmpTweets:
            if tweet.created_at < endDate and tweet.created_at > startDate:
                tweets.append(tweet)
        # print(tmpTweets[-1].created_at, startDate)

    print("Stopping dates")
    print(tmpTweets[-1].created_at, startDate)
    # tweets
    
    column_names = ["tweet_id", "tweet_created_at", "tweet_text", "hashtags" ]
    df = pd.DataFrame(columns = column_names)

    for idx,tweet in enumerate(tweets):
        tweet_id = str(tweet.id)
        tweet_created_at = str(tweet.created_at)
        tweet_text = tweet.text
        hashtags = ''
        try:
            a = tweet.entities
            b = (a['hashtags'][0])
            # hashtags = b
            # print(b)
            # print(b['text'])
            hashtags = b['text']
        except:
            hashtags = ''
        row = [tweet_id, tweet_created_at, tweet_text, hashtags]
        df.loc[len(df)] = row


    directory = file_directory+country
    if not os.path.exists(directory):
        os.makedirs(directory)
    df.to_csv(directory+'/'+handle_name+'.csv', index = False)
```

```
# get_tweets("usagov", "usa")
country = "USA"
# "usagov",
handle_list = [ "realDonaldTrump", "Mike_Pence", "SecAzar", "SecPompeo"]

for each_handle in handle_list:
    get_tweets(each_handle, country)

print("done")
```
