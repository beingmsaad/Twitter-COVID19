```
# Define the web scraping function
def webcapture(url):
    headers = {
   # pretend I am a browser
   'User-Agent': 'Mozilla/5.0',
   }
    session = requests.Session() #setup session
    data = session.get(url, headers=headers) #scrape the data
    soup = BeautifulSoup(data.text, 'html.parser') #parse the data
    return soup #return the parsed data
```

```
sdate = dt.date(2019, 12, 15)   # start date
# edate = dt.date(2020, 4, 20)   # end date
edate = dt.date(2019, 12, 16)   # end date

delta = edate - sdate       # as timedelta
print(type(delta))
date_range = []
for i in range(delta.days + 1):
    day = sdate + dt.timedelta(days=i)
    date_range.append(str(day))
# print(date_range)
```

```
#old get trend
def old_get_trend(country_name, country_id):

    # Scrape the website
    url_base = 'https://trendogate.com/placebydate/'
    usa_trendings = pd.DataFrame()
    # date_range1 = ['2020-1-1']

    column_names = ["date", "trend_1", "trend_2", "trend_3", "trend_4", "trend_5" ]
    df = pd.DataFrame(columns = column_names)

    for idx,date in tqdm(enumerate(date_range)):
        url = url_base+country_id+'/'+date
        soup = webcapture(url)
        trending_list = [li.get_text() for li in soup.find_all('li','list-group-item')]
        
        trend_1 = ""
        trend_2 = ""
        trend_3 = ""
        trend_4 = ""
        trend_5 = ""

        try:
            trend_1 = trending_list[0]
        except:
            trend_1 = ""

        try:
            trend_2 = trending_list[1]
        except:
            trend_2 = "" 

        try:
            trend_3 = trending_list[2]
        except:
            trend_3 = ""
        try:
            trend_4 = trending_list[3]
        except:
            trend_4 = ""
        try:
            trend_5 = trending_list[5]
        except:
            trend_5 = ""
    
        row = [date, trend_1, trend_2, trend_3, trend_4, trend_5]
        df.loc[len(df)] = row
        time.sleep(1)

    # print(df.head())
    directory = file_directory
    df.to_csv(directory+'/'+country_name+'.csv', index = False)

    # return df.head(1)
```

```
#new get_trend
def get_trend(country_name, country_id):
    # country_id = '23424977'
    # country_name = 'USA'
    # Scrape the website
    url_base = 'https://trendogate.com/placebydate/'
    usa_trendings = pd.DataFrame()
    # date_range1 = ['2020-1-1']

    column_names = ["date", "trends"]
    df = pd.DataFrame(columns = column_names)

    for idx,date in tqdm(enumerate(date_range)):
        url = url_base+country_id+'/'+date
        soup = webcapture(url)
        trending_list = [li.get_text() for li in soup.find_all('li','list-group-item')]
        
        trend_list = []
        for each_trend in trending_list:
            if('#' in each_trend):
                trend_list.append(each_trend)
        # print(date + " -> "+ str(trend_list))
        row = [date, trend_list]
        # print(row)
        df.loc[len(df)] = row
        time.sleep(1)

    # print(df.head())
    directory = file_directory
    df.to_csv(directory+'/'+country_name+'.csv', index = False)

    # return df.head()
```
